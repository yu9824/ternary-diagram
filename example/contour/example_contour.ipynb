## import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ternary_diagram import TernaryDiagram
```

## load example data for contour

```python
df = pd.read_csv("example_contour.csv")
df.head()
```

Compounds corresponding to vertices

Specify triangle labels around the half-hour mark.

```python
materials = df.columns[0:3].tolist()
materials
```

## Easiest example

```python
td = TernaryDiagram(materials=materials)
td.contour(df[materials], z=df["z"])
```

## customize

### colorbar label

```python
td = TernaryDiagram(materials=materials)
mappable = td.contour(df.iloc[:, :3], df.loc[:, "z"], flag_cbar=False)
# you can customize the colorbar like `plt.colorbar`
td.colorbar(mappable=mappable, label="Example of `contour`")
```

```python
td.fig.savefig("example_contour.png", dpi=144)
```

```python
td = TernaryDiagram(materials=materials)

mappable = td.contour(
    df.loc[:, materials], df.loc[:, "z"].round(1), flag_cbar=False, fill=False
)
# # you can customize the colorbar like `plt.colorbar`
td.colorbar(mappable=mappable, label="Example of `contour`")
```

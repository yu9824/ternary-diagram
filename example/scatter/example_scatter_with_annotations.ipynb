```python
# `element_recognition` is a module for recognizing elements from chemical formula.
# See: https://github.com/yu9824/element_recognition
import element_recognition
import matplotlib.pyplot as plt

from ternary_diagram import TernaryDiagram

element_recognition.__version__
```

```python
materials = ["Li4SiO4", "Li4GeO4", "Li4SnO4"]
products = ["Li4Si0.1Ge0.4Sn0.5O4"]
```

```python
ratio = element_recognition.get_ratio(materials=materials, products=products)
ratio
```

```python
fig, ax = plt.subplots(facecolor="white", dpi=144)
td = TernaryDiagram(materials=materials, ax=ax)
```

```python
# you can annotate the points with list of strings
td.scatter(ratio, annotations=ratio.index)
fig
```
